```python
import pandas as pd

df = pd.read_csv('spotify.csv')  # 替换成你的文件名
df.head()  # 显示前5行
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style( 'whitegrid')
plt. rcParams ['figure.figsize'] = (10, 6)
```

```python
mean_age = df['age'].mean()
median_age = df['age'].median()
mode_age = df['age'].mode()

print(f"Mean age: {mean_age}")
print(f"Median age: {median_age}")
print(f"Mode age: {mode_age.tolist()}") # mode() can return multiple values, so convert to list for display
```

```python
country_counts = df['country'].value_counts()

plt.figure(figsize=(12, 6))
ax = sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title('Distribution of Users by Country')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)

# Add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
subscription_counts = df['subscription_type'].value_counts()
subscription_proportions = df['subscription_type'].value_counts(normalize=True) * 100

print("Subscription Type Counts:")
print(subscription_counts)
print("\nSubscription Type Proportions (%):")
print(subscription_proportions)
```

```python
subscription_counts = df['subscription_type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(subscription_counts, labels=subscription_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Subscription Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
